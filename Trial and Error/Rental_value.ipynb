```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
import numpy as np
import holoviews as hv
```

```python
rental_data_path = Path('/Users/yuvrajkabra/Documents/Bootcamp/Project 1/Real_Estate_Project/Rental_data.csv')
rental_data_df = pd.read_csv(rental_data_path).dropna()
```

```python
rental_data = rental_data_df.drop(columns=["RegionID", "SizeRank"])
rental_data
```

```python
rental_data['RegionName'] = rental_data['RegionName'].str.split(',').str.get(1)
```

```python

rental_data
```

```python
rental_data['RegionName'] = rental_data['RegionName'].drop_duplicates(keep='first')
```

```python
rental_data = rental_data.dropna()
```

```python
rental_data
```

```python
rental_data = rental_data.transpose()
```

```python
rental_data.columns = rental_data.iloc[0]
rental_data = rental_data[1:]
rental_data.head()
```

```python
rental_data.hvplot(
    ylabel='Rental Income',
    title="Average US Rental Income By State ",
    ).opts(yformatter='%.0f')
```

```python
rental_each_state = rental_data.hvplot(
    ylabel='Rental Income',
    title="Average US Rental Income By State ",
    ).opts(yformatter='%.0f')
individual_plots = hv.HoloMap(rental_each_state).opts(yformatter='%.0f')
individual_plots
```
