```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
import numpy as np
import holoviews as hv
import sqlalchemy as sql
```

```python
rental_data_path = Path('/Users/yuvrajkabra/Documents/Bootcamp/Project 1/Real_Estate_Project/Metro_ZORI_AllHomesPlusMultifamily_Smoothed.csv')
rental_data_df = pd.read_csv(rental_data_path).dropna()
rental_data_df.head()
```

```python
rental_data_df[['City', 'State']] = rental_data_df['RegionName'].str.split(',', expand=True) 
rental_data_df.head()
```

```python
rental_data_df = rental_data_df.sort_values('2022-06', ascending=False)
rental_data_df.head(10)
```

```python
top_rent_cities = rental_data_df.head(10)
top_rent_cities
```

```python
top_rent_cities = top_rent_cities.drop(columns=["RegionID", "RegionName", "SizeRank", "State"])
top_rent_cities.reset_index()
```

```python
top_rent_cities['City'] = top_rent_cities['City'].str.replace('Miami-Fort Lauderdale', 'Miami')
top_rent_cities['City'] = top_rent_cities['City'].str.replace('Los Angeles-Long Beach-Anaheim', 'Los Angeles')
top_rent_cities['City'] = top_rent_cities['City'].str.replace('Urban Honolulu', 'Honolulu')
```

```python
top_rent_cities = top_rent_cities.set_index('City')
top_rent_cities
```

```python
top_rent_cities_plot = top_rent_cities.transpose()
top_rent_cities_plot
```

```python
top_rent_cities_plot.hvplot(
    xlabel="Date",
    ylabel="Average Rent",
    title="Average Rent Price By City",
    ).opts(yformatter='%.0f')
```

```python
rent_each_city = top_rent_cities_plot.hvplot(xlabel="Date", ylabel="Average Rent", title="Average Rent Price By City").opts(yformatter='%.0f')
individual_plots = hv.HoloMap(rent_each_city).opts(yformatter='%.0f')
individual_plots
```

```python
housing_data_path = Path("/Users/yuvrajkabra/Documents/Bootcamp/Project 1/Real_Estate_Project/City_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_month.csv")
housing_data_df = pd.read_csv(housing_data_path)
housing_data_df.head()
```

```python
database_connection_string = 'sqlite:///'
engine = sql.create_engine(database_connection_string)
housing_data_df.to_sql('property_value', engine, index=True, if_exists='replace')
engine.table_names()
```

```python
query = f"""
SELECT *
FROM property_value
WHERE RegionName in ('San Jose', 'San Francisco', 'New York', 'San Diego', 'Ventura', 'Los Angeles', 'Miami',
'Boston', 'Honolulu', 'Stamford')
"""
```

```python
housing_df = pd.read_sql_query(query, engine)
housing_df.head()
```

```python
housing_df = housing_df.groupby("RegionName").mean()
housing_df = housing_df.drop(columns=["RegionID", "SizeRank", "index"])
housing_df.index.names = ['City']
```

```python
housing_df
```

```python
housing_df_plot = housing_df.transpose()
housing_df_plot
```

```python
housing_df_plot.hvplot(
    xlabel="Date",
    ylabel="Average Home Value",
    title="Average Property Value By City",
    ).opts(yformatter='%.0f')
```

```python
housing_each_city = housing_df_plot.hvplot(xlabel="Date", ylabel="Average Home Value", title="Average Property Value By City").opts(yformatter='%.0f')
ind_housing_plots = hv.HoloMap(housing_each_city).opts(yformatter='%.0f')
ind_housing_plots
```
